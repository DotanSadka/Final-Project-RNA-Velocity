```python
import pandas as pd
```

load the csv files (the counts 0)

```python
csv_files = [
    'counts_0_N_100.csv',
    'counts_0_N_313.csv',
    'counts_0_N_364.csv',
    'counts_0_N_880.csv',
    'counts_0_N_982.csv'
]
```

turn it into a datafram

```python
dfs = [pd.read_csv(file, index_col=0).rename_axis('cell_name').reset_index() for file in csv_files]
```

combine all the matrix into one and saving it as a csv file

```python
combined_df = pd.concat(dfs, ignore_index = True, sort = False).fillna(0)

combined_df.to_csv('combined_counts_0.csv', index = False)
```
